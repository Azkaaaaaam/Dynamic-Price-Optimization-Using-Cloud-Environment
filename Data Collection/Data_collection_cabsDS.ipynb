```python
import requests
import pandas as pd
from datetime import datetime
from google.colab import files
import matplotlib.pyplot as plt
```

```python
def retrieve_data(url, skip=0, top=1000, step=3000, filter=None):
    params = {'$skip': skip, '$top': top}
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print("Failed to retrieve data")
        return None
```

```python
url = "https://data.cityofnewyork.us/api/odata/v4/m6nq-qud6"

all_data = []
skip = 0
step = 3000
```

```python

while len(all_data) <= 3000000:

    data = retrieve_data(url, skip=skip, top=1000)
    if data is None:
        break
    all_data.extend(data['value'])
    skip += step
    if len(data['value']) < 1000:
        break
```

```python
df = pd.DataFrame(all_data)
```
